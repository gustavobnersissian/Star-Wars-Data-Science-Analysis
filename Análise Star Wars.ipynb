```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
characters = pd.read_csv("./Inputs/characters.csv")
planets = pd.read_csv("./Inputs/planets.csv")
species = pd.read_csv("./Inputs/species.csv")
starships = pd.read_csv("./Inputs/starships.csv")
vehocles = pd.read_csv("./Inputs/vehicles.csv")
```

## Gênero

### Total de personagens por gênero

```python
generos = characters['gender'].value_counts().to_frame().reset_index()
generos.columns = ['Gênero', 'Total'] 
generos
```

# Altura

```python
# Top 10 personagens mais altos
top_10_altura = characters.sort_values(by=['height'], ascending=False).head(10).reset_index()
top_10_altura
```

### Média de altura

```python
media_altura = characters['height'].mean()
media_altura = round(media_altura, 2)
media_altura
print("Média de altura (cm) = ", media_altura)
```

### Mediana de altura

```python
mediana_altura = characters['height'].median()
mediana_altura = round(mediana_altura, 2)
mediana_altura
print("Mediana de altura (cm) = ", mediana_altura)
```

### Moda de altura

```python
moda_altura = characters['height'].mode()
moda_altura = round(moda_altura, 2)
print("Moda de altura (cm) = ", moda_altura[0])
```

## Comparações com altura

### Top 10 personagens mais altos

```python
ax = sns.barplot(x="name", y="height", data = top_10_altura)
ax.figure.set_size_inches(25, 10)
ax.set_title('Comparação de Altura (cm) - Top 10 mais altos', fontsize = 30)
ax.set_xlabel('Nome', fontsize = 30)
ax.set_ylabel('Altura (cm)', fontsize = 30)             
```

```python
# Top 10 personagens mais baixos
top_10_altura_baixos = characters.sort_values(by=['height'], ascending=True).head(10).reset_index()
top_10_altura_baixos = top_10_altura_baixos.sort_values(by=['height'], ascending=False)
top_10_altura_baixos
```

### Top 10 personagens mais baixos

```python
ax = sns.barplot(x="name", y="height", data = top_10_altura_baixos)
ax.figure.set_size_inches(31, 10)
ax.set_title('Comparação de Altura (cm) - Top 10 mais baixos', fontsize = 35)
ax.set_xlabel('Nome', fontsize = 30)
ax.set_ylabel('Altura (cm)', fontsize = 30)     
```

### Distribuição das frequências - Altura

```python
ax = sns.distplot(characters['height'],
                  kde = True # função de densidade
)
sns.set(font_scale = 2)
ax.set_title("Distribuição de Frêquencias - Altura", fontsize = 20)
ax.figure.set_size_inches(18, 6)
ax.set_xlabel('(cm)', fontsize = 17)
ax.set_ylabel('(%)', fontsize = 17)
```

### Boxplot  - Altura 
- Obs: conseguimos observar através deste gráfico, onde o valor está mais concentrado, os quartis e também os outliers.
- Os pontinhos pretos representam os outliers, valores muito altos, ou baixos, que não estão presentes onde os valores estão mais concentrados (retângulo azul).

```python
ax = sns.boxplot(x = 'height', data = characters, orient = 'y')
ax.figure.set_size_inches(12, 4)
ax.set_title('Altura', fontsize=25)
ax.set_xlabel('(cm)', fontsize=20)
ax
```

```python
quartis_personagens = characters['height'].quantile([0.25, 0.5, 0.75])
quartis_personagens
```

### Quartis - Altura
- Obs: Os quartis dividem os dados em partes iguais. Com eles você pode rapidamente avaliar a dispersão e a tendência central de um conjunto de dados, que são etapas importantes na compreensão dos seus dados.
- Como observamos na tabela a baixo, 25% dos valores são menores ou iguais a 167cm. 
- 50% dos valores são menores ou iguais a 180cm. 
- E 75% dos valores são menores ou iguais a 191cm.

```python
quartis_personagens_df = pd.DataFrame(quartis_personagens)
quartis_personagens_df.rename(index={0:'Quartis'}, columns={1:'Altura'}, inplace=True)
quartis_personagens_df = quartis_personagens_df.reset_index()
quartis_personagens_df
quartis_personagens_df.rename(columns={'index':'Quartis'}, inplace=True)
quartis_personagens_df
```
